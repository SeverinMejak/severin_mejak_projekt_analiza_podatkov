##Opis projekta

V In [37]: vidimo koliko člankov je vključeno v analizi po temah. V uvodu sem povedal, da bom iz vsake teme vzel po 20 člankov. Članke sem jemal s strani http://www.rtvslo.si/{0}/arhiv, kjer sem {0} formatiral z temo. Vsak članek ima url, ki vključuje natanko eno temo. Tako se lahko članek nanaša na več tem, a ima v url-ju samo eno. Ker nisem želel, da se članki ponavljajo, sem zahteval, da ima članek, ki sem ga vzel iz arhiva neke teme tudi to temo v url-ju. Ker je stran www.rtvslo.si narejena tako, da ima na vrhu strani in na desni strani še povezave do najbolj popularnih in najbolje ocenjenih člankov, je v resnici na strani od arhiva več kot 20 člankov. Če upoštevam kriterij omenjen zgoraj, nanese tako, da je člankov pri nekaterih temah več kot 20, pri nekaterih pa manj. To ne bo bistveno uplivalo na rezultat, saj bom gledal povprečje.

```python
tabela[tabela.ocena_novice != 0.0].groupby('tema').mean().sort_values('ocena_novice', ascending=False)[['ocena_novice']]
```

Iz zgornje tabele vidimo, da ima najboljšo povprečno oceno tema znanost in tehnologija, najnižjo pa Slovenija. To, da so znanost in tehnologija, šport in tureavanture najvišje ocenjene, sem pričakoval v hipotezi. Prav tako sem pričakoval, da bosta zabava in svet slabo ocenjena. Malo pa me je presenetilo, da je Slovenija ocenjena naslabše. To morda pomeni, da nekateri negativno ocenijo novice zgolj zato, ker nenegativno opisujejo kakega politika, ki ga ne marajo.

Opomba: Najprej sem izločil tiste članke, ki so imeli povprečno oceno 0 (to pomeni, da še niso bili ocenjeni, saj je ocena lahko med 1 in 5).

```python
tabela.groupby('tema').mean().sort_values(['stevilo_komentarjev', 'stevilo_ocen'], ascending=[False, False])[['stevilo_komentarjev', 'stevilo_ocen']]
```

Z mojo hipotezo sem precej zadel; res je najbolj popularna tema slovenija, ter nato šport in svet. Tudi za zdravje se mi je dozdevalo, da nam je pomembno. Prav tako sem imel na žalost prav, saj je kultura najmanj popularna (daleč najmanj popularna). Preseneča me to, da zabava ni bolj poularna in da je tema ture avanture manj.

```python
tabela['vseh_ocen'] = tabela.stevilo_pozitivnih + tabela.stevilo_negativnih
```

```python
tabela2 = tabela[tabela.vseh_ocen != 0].groupby('tema').mean()[['stevilo_pozitivnih', 'stevilo_negativnih', 'vseh_ocen']]
```

```python
tabela2
```

```python
tabela2['relativna_razlika'] = (tabela2.stevilo_pozitivnih - tabela2.stevilo_negativnih)/ tabela2.vseh_ocen
```

```python
tabela2.sort_values('relativna_razlika', ascending= False)
```
